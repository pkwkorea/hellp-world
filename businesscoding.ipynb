```python
# 단리 정기예금 원리금과 이자 계산

def main():
    
    interest_rate = float(input('이자율을 소숫점으로 입력: '))
    years = int(input('만기를 연으로 입력: '))
    principal = int(input('저축원금 입력: '))
    total=interest(principal, years, interest_rate)
    print("원리금:", format(total, ",.0f"))

def interest(deposit, maturity, rate):
    before_tax_sum = deposit * (1+rate *maturity)
    return before_tax_sum

main()
```

```python
# 만기별 이자율을 반영한 정기예금 원리금 계산

def main():
    principal=int(input("원금 입력;"))
    maturity = float(input('만기 년수 입력'))
    rate = calc_rate(maturity)
    before_tax_sum = principal *(1+rate*maturity) 
    print('세전 원리금; ', format(before_tax_sum, ',.2f'), sep='')

def calc_rate(x):
    if x < 0.25:
        rate = 0.0135
    elif x >= 0.25 and x < 0.5:
        rate = 0.0145
    elif x >= 0.5 and x < 1:
        rate = 0.0155
    elif x >=1  and x <2:
        rate = 0.0165
    elif x >=2  and x <3:
        rate = 0.0175
    else:
        rate = 0.0185
    return rate

main()
```

```python
#월복리 적금 원리금 계산

# A=installment r=interest rate n=maturity m=number of installments per year

import numpy as np

def main():
    
    A=int(input("월적립액 입력:"))
    r=float(input("연이자율 입력:"))
    m=int(input("연간 적립횟수 입력:"))
    n=int(input("만기년수 입력:"))
    taxrate=float(input("이자소득세율 입력:"))
    
    CF=np.ones((1,n*m))*A
    exponents=np.array(range(1,n*m+1))
    fvfs=(1+r/m)**exponents
    fv_installment=np.sum(CF*fvfs)
    after_tax=fv_installment-(fv_installment-A*m*n)*taxrate
    
    print("월복리 적금의 세전 원리금은",format(fv_installment,",.0f"), "입니다")
    print("월복리 적금의 세후 원리금은",format(after_tax,",.0f"), "입니다")
    
main()
```

```python
# 목표 금액 마련을 위한 월 적립액 계산(월복리 정기예금)

from sympy.solvers import solve
from sympy import Symbol

target=int(input("목표금액을 입력;"))
r=float(input("이자율을 입력;"))
m=int(input("연 적립횟수를 입력;"))
n=int(input("만기년도를 입력;"))
taxrate=float(input("이자소득세율 입력:"))
A=Symbol("A")
    
CF=np.ones((1,n*m))*A
exponents=np.array(range(1,n*m+1))
fvfs=(1+r/m)**exponents
fv_installment=np.sum(CF*fvfs)
after_tax=fv_installment-(fv_installment-A*m*n)*taxrate
sol=solve(after_tax-target,A)  # Warning: sol is returned as a list

sol=sol[0]  
print("세후 목표금액 마련을 위한 월 적립액:", format(sol, ",.0f"))
```

```python
# 단리 적금 원리금 계산

class Install:
    
    def __init__(self, rate,maturity, amount):
        self.rate = rate
        self.maturity = maturity
        self.amount = amount
    
    def total(self):
        before_tax_sum=(self.amount*(12*self.maturity)*(2+(12*self.maturity+1)*(self.rate/12)))/2
        return before_tax_sum
    
class Install_sub(Install):
    def __init__(self,rate,maturity, amount):
        super().__init__(rate,maturity, amount)

    def interest(self):
        totalamount=super().total()
        interest=totalamount-(self.amount*12*self.maturity)
        return interest
    
def main():
    rate = float(input ("연이자율 입력: "))
    maturity = int(input("만기년수 입력: "))
    amount = int(input( "월적립액 입력: "))
    y = Install_sub(rate, maturity, amount) 
    print("세전원리금;", format(y.total(), ',.2f'))
    print("총이자액;", format(y.interest(), ',.2f'))

main()
```

```python
# 연금현재가치 계산(복리)

# A=payment r=period rate n=number of total payments m=number of payments per year

import numpy as np

def main():
    
    A=int(input("연금액 입력:"))
    r=float(input("이자율 입력:"))
    m=int(input("연간 연금횟수 입력:"))
    n=int(input("총 연금횟수 입력:"))

    
    CF=np.ones((1,n))*A
    exponents=np.array(range(1,n+1))
    fvfs=(1+r/m)**exponents
    pvfs=np.reciprocal(fvfs)
    pv_annuity=np.sum(CF*pvfs)
    print("연금의 현재가치는",format(pv_annuity,",.0f"), "입니다")
    
main()
```

## 이표채 가격

```python
# This is a python function for calculating price of copon bond
# face=par value                couponrate=coupon rate
# d= number of days to first coupon from today   
# q= coupon interval(days) 
# n=number of remaining coupons   m=coupon frequency per year
# yield=bond yield

import numpy as np

def couponprice(face,couponrate,d,q,n,m,bondyield):

    l=np.ones((1,n))
    coupon=face*(couponrate/m)
    coupons=coupon*l
    periodyield=1+(bondyield/m)
    exponents=(d/q)+np.array(range(n))
    pvfs=np.reciprocal(periodyield**exponents)
    price=np.sum(coupons*pvfs)+face*pvfs[n-1]

    return price 
```

## 대출상환 스케쥴

```python
# 원리금균등분할상환 상환 스케쥴

def main():
    
    principal=int(input("대출원금 입력:"))
    rate=float(input("이자율 입력:"))
    maturity=int(input("만기년수 입력:"))

    payment = (principal * (rate/12)) /(1 - (1 / (1 + (rate/12)) ** (maturity * 12)))
    print("회차","\t" ,"기초원금","\t" ,"월상환액","\t" ,"이자","\t","\t","상환원금","\t" ,"기말원금")
    print("------------------------------------------------------------------------------------")
    
    begBal = principal

    for num in range(1, maturity*12+1):
        interest = begBal * (rate /12)
        principalpay = payment - interest
        endBal = begBal - principalpay

        print(num,"\t" , format(begBal, "8,.0f"), "\t" ,format(payment, "8,.0f"),"\t" , \
              format(interest, "8,.0f"),"\t" , format(principalpay, "8,.0f"),"\t" , format(endBal, "8,.0f"))
        begBal = endBal
    
main()
```

```python
# 원리금 균등분할상환

class Loan:
    
    def __init__(self, rate,maturity, principal):
        self.rate = rate
        self.maturity = maturity
        self.principal = principal
    
    def monthlypay(self):
        payment = (self.principal * (self.rate/12)) /\
        (1 - (1 / (1 + (self.rate/12)) ** (self.maturity * 12)))
        return payment
    
    def interest(self):
        firstinterest= (self.principal * (self.rate/12)) 
        return firstinterest   
   
    def principalpay(self):
        principalpay= self.monthlypay() -self.interest()
        return principalpay
    
    def balance(self):
        balance=self.principal-self.interest()
        return balance
    
    def totalinterest(self):
        totalinterest=self.monthlypay()*self.maturity*12-self.principal
        return totalinterest

def main():
    
    rate = float(input("연이자율 입력:"))
    maturity = int(input("만기년수 입력:"))
    principal = int(input("원금 입력:"))

    # 객체 생성
    x = Loan(rate,maturity,principal)
  
    print()
    print("월상환금은", format(x.monthlypay(), '6,.0f'), "입니다.")
    print("1회차이자는", format(x.interest(), '6,.0f'), "입니다.")
    print("1회차원금상환액은", format(x.principalpay(), '6,.0f'), "입니다.")
    print("상환후원금잔액은", format(x.balance(), '6,.0f'), "입니다.")
    print("총이자 부담액은", format(x.totalinterest(), '6,.0f'), "입니다.")

main()
```

```python
# 회차별 할부 상환액 계산

class Creditcard:
    
    def __init__(self,months, amount):
        self.months = months
        self.amount = amount
    
    def monthlypay(self):
        monthlypay = self.amount/self.months
        return monthlypay
    
class Creditcard_1(Creditcard):
    def __init__(self,months, amount,rate,remainingmonths):
        super().__init__(months, amount)
        self.rate = rate
        self.remainingmonths=remainingmonths
        
    def monthlypay(self):
        principal=super().monthlypay()
        interest=principal*(self.remainingmonths)*(self.rate/12)
        totalpay=principal+interest
        return totalpay
def main():
    
    months = int(input ("할부개월수 입력: "))
    amount = int(input("카드사용액입력: "))
    rate = float(input( "할부수수료율을 입력: "))
    remainingmonths = int(input( "잔존회차 입력: "))
    y = Creditcard_1(months, amount,rate,remainingmonths)

    if isinstance(y, Creditcard_1):
        print("다음 결제일 총상환금은", format(y.monthlypay(), ',.2f'), "원입니다.")
    else:
        print('wrong object')
```

```python
import pandas as pd
import numpy as np

np.random.seed(10)
sales_mu, sales_sigma= 3000, 100
inv_mu, inv_sigma=200, 10
receiv_mu, receiv_sigma=400, 11
pay_mu, pay_sigma=120, 9

np.random.seed(100)
M=50000
sales=pd.DataFrame(np.random.normal(sales_mu,sales_sigma , M))
inv=pd.DataFrame(np.random.normal(inv_mu, inv_sigma, M))
receiv=pd.DataFrame(np.random.normal(receiv_mu,receiv_sigma , M))
pay=pd.DataFrame(np.random.normal(pay_mu,pay_sigma , M))

simul=pd.concat([sales,inv], axis=1)
simul=pd.concat([simul, receiv],axis=1)
simul=pd.concat([simul, pay],axis=1)

simul.columns=('sales','inventory','receivables','payables')
simul["NWC"]=simul['inventory']+simul['receivables']-simul['payables']
simul['inv_days']=365/(simul['sales']/simul['inventory'])
simul['receiv_days']=365/(simul['sales']/simul['receivables'])
simul['OC']=simul['inv_days']+simul['receiv_days']
simul.head()
```

```python
import pandas as pd
import numpy as np

# fcff simulation

np.random.seed(10)
NI_now=2338
dep_now=1300
receiv_now=3906
inv_now=1961
pay_now=1268
NWC_now=receiv_now+inv_now-pay_now

NI_mu, NI_sigma=NI_now, 400
receiv_mu, receiv_sigma=receiv_now, 200
inv_mu, inv_sigma=inv_now, 100
pay_mu, pay_sigma=pay_now, 50

M=50000

NI=pd.DataFrame(np.random.normal(NI_mu,NI_sigma , M))
receiv=pd.DataFrame(np.random.normal(receiv_mu,receiv_sigma , M))
inv=pd.DataFrame(np.random.normal(inv_mu, inv_sigma, M))
pay=pd.DataFrame(np.random.normal(pay_mu,pay_sigma , M))

simul=pd.concat([NI,receiv], axis=1)
simul=pd.concat([simul, inv],axis=1)
simul=pd.concat([simul, pay],axis=1)

simul.columns=('NI_est','receivables_est', 'inventory_est','payables_est')
simul["NWC_est"]=simul['receivables_est']+simul['inventory_est']-simul['payables_est']
simul["NWCadd"]=simul['NWC_est']-NWC_now
simul["fcff_est"]=simul['NI_est']+dep_now-simul["NWCadd"]

# 기말 현금보유액 simulation
cash_now=1103
simul["cash_est"]=simul['fcff_est']+cash_now
simul.sample(5)
```

```python
simul['cash_group']=pd.cut(simul["cash_est"],bins=70)
table=pd.crosstab(index=simul["cash_group"],columns='빈도')
table["확률"]=table["빈도"]/M
table["누적확률"]=table["확률"].cumsum()
table
```
