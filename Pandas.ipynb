# melt

```python
import pandas as pd
cheese = pd.DataFrame({'first' : ['John', 'Mary'],'last' : ['Doe', 'Bo'], 'height' : [5.5, 6.0], 'weight' : [130, 150]})
print("original data;")
print(cheese)

#성과 이름별로 height, weight를 표시
cheese1=cheese.melt(id_vars=['first', 'last'])
print(cheese1)

# variable에 변수 이름 지정
cheese2=cheese.melt(id_vars=['first', 'last'],var_name='specs')
print(cheese2)
```

```python
import pandas as pd
roster=pd.read_excel('roster.xlsx')
roster.head()
```

```python
def grade(x):
    if x>=85:
        grade= "A+"
    elif x>=76:
         grade= "A"
    elif x>=65:
        grade= "B+"
    elif x>=50:
        grade= "B"
    elif x>=40:
        grade= "C+"
    elif x>=30:
        grade= "C"
    elif x>=20:
        grade= "D+"
    elif x>=10:
        grade= "D"
    else:
        grade= "F"
    return grade
    
roster['학점']=roster["평균"].apply(grade) 
roster
```

```python
학점별인원=pd.crosstab(index=roster['학점'],columns='인원')
학점별인원
```

```python
newid=roster["사용자 ID"].astype(str)
roster['사용자 ID']=newid

def hideid(x):
    hideid=x[5:]
    return hideid

roster['사용자 ID']=roster["사용자 ID"].apply(hideid) 
roster
```

# monte carlo simulation

```python
import pandas as pd
import numpy as np

np.random.seed(10)
mainmu, mainsigma= 15,3
rawmu, rawsigma=6,2
prodmu, prodsigma=25000, 7000

np.random.seed(100)
M=50000
main=pd.DataFrame(np.random.normal(mainmu, mainsigma, M))
raw=pd.DataFrame(np.random.normal(rawmu, rawsigma, M))
prod=pd.DataFrame(np.random.normal(prodmu, prodsigma, M))
simul1=pd.concat([main,raw], axis=1)
simul=pd.concat([simul1, prod],axis=1)
simul.columns=('mainsaving','rawsaving','production')
simul['costsaving']=(simul['mainsaving']+simul['rawsaving'])*simul['production']

simul['savinggroup']=pd.cut(simul["costsaving"],bins=100)
table=pd.crosstab(index=simul["savinggroup"],columns='빈도')
table["확률"]=table["빈도"]/M
table["누적확률"]=table["확률"].cumsum()
table
```

```python
%matplotlib tk
simul['costsaving'].plot.hist(100)
```

```python
# EAR 계산

import numpy as np
import pandas as pd
%matplotlib tk

np.random.seed(10)
M=30000
ear=pd.DataFrame(np.random.normal(1080, 25, (M,4)))
ear.columns=['Q1rate','Q2rate','Q3rate','Q4rate']
ear["revenue"]=ear['Q1rate']*20000+ear['Q1rate']*18000+ear['Q1rate']*20000+ear['Q1rate']*24000
ear['NP']=ear["revenue"]*0.4*(1-0.2)
ear['NPgroup']=pd.cut(ear["NP"],bins=100)
table=pd.crosstab(index=ear["NPgroup"],columns='빈도')
table["확률"]=table["빈도"]/M
table["누적확률"]=table["확률"].cumsum()
table
```

```python
%matplotlib tk
ear['NP'].plot.hist(100)
```
